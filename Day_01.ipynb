## 6. What is a marker?
A marker is a symbol used to represent individual data points on a plot.

```python
# Example of a marker

import matplotlib.pyplot as plt

x = [10, 20, 34]
y = [1, 2, 3]

fig = plt.plot(x, y, marker='D', markerfacecolor = 'red', markeredgecolor = 'black')  # Creates diamond-shaped markers at each data point (x, y)

```

## 8. What is line style?
In matplotlib, linestyle controls the appearance of the line that connects your data points.
It defines how the line looks â€” solid, dashed, dotted, etc.
ðŸ§ª Common linestyle Options
| Symbol | Style Name | Description | 

| '-' | Solid line | Default continuous line | 

| '--' | Dashed line | Line with dashes | 

| '-.' | Dash-dot line | Dash followed by a dot | 
    
| ':' | Dotted line | Line made of dots | 


```python
x = [10, 20, 34]
y = [1, 2, 3]

plt.plot(x, y, marker='D', markerfacecolor = 'red', markeredgecolor = 'black', linestyle = '--')  # Creates diamond-shaped markers at each data point (x, y)


plt.show()
```

## 9. What is legend?
A legend in matplotlib is a small box that explains what each plotted line, marker, or shape represents.
It maps labels to visual elements in your plot â€” like colors, line styles, or markers â€” so viewers can understand what theyâ€™re looking at.


```python
x = [10, 20, 34]
y = [1, 2, 3]
plt.xlabel('Value of x')
plt.ylabel('Value of y')
plt.plot(x, y, marker='D', markerfacecolor = 'red', markeredgecolor = 'black', linestyle = ':', color='gray', label='My data series')  # Creates diamond-shaped markers at each data point (x, y)

plt.legend(title="This is a legend", loc='upper center')

plt.show()
```
